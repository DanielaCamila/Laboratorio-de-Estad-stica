### Marco Teórico


#### Regresión Logística
La regresión logística es un algoritmo de aprendizaje supervisado utilizado para tareas de clasificación binaria. Su objetivo es predecir la probabilidad de que una instancia pertenezca a una de dos clases posibles. A diferencia de la regresión lineal, que modela relaciones continuas, la regresión logística aplica la función logística (sigmoide) para mapear cualquier valor real en un rango entre 0 y 1, facilitando así la interpretación probabilística de las predicciones. Este modelo es ampliamente utilizado en diversas áreas, como la medicina y las ciencias sociales, para predecir la ocurrencia de eventos discretos. [1]


#### Máquinas de Vectores de Soporte (SVM)
Las Máquinas de Vectores de Soporte (Support Vector Machines, SVM) son algoritmos de aprendizaje supervisado utilizados tanto para tareas de clasificación como de regresión. El principio fundamental de las SVM es encontrar el hiperplano óptimo que separa las clases en un espacio de características, maximizando el margen entre los puntos de datos más cercanos de cada clase, conocidos como vectores de soporte. Para manejar datos no linealmente separables, las SVM emplean funciones núcleo (kernels) que transforman el espacio de características original en uno de mayor dimensión donde la separación lineal es posible. [2]

#### Redes Neuronales (Perceptrón Multicapa)
Un Perceptrón Multicapa (Multi-Layer Perceptron, MLP) es un tipo de red neuronal artificial compuesta por al menos tres capas de nodos: una capa de entrada, una o más capas ocultas y una capa de salida. Cada neurona en una capa está conectada a todas las neuronas de la siguiente capa, formando una estructura completamente conectada. Los MLP utilizan funciones de activación no lineales, como la función sigmoide o ReLU, permitiendo que la red aprenda representaciones complejas y capture relaciones no lineales en los datos. Son ampliamente utilizados en tareas de clasificación y regresión debido a su capacidad para modelar patrones complejos. [3]

#### Kernels
En el contexto del aprendizaje automático, una función núcleo (kernel) es una técnica que permite transformar datos en un espacio de características de mayor dimensión sin necesidad de calcular explícitamente las nuevas coordenadas. Esto se logra mediante el cálculo de productos internos en el espacio transformado, facilitando la aplicación de algoritmos lineales a problemas que son no linealmente separables en su espacio original. El uso de kernels es fundamental en algoritmos como las SVM, donde permiten manejar datos complejos y encontrar hiperplanos de separación en espacios de alta dimensión. [4]

#### Métricas para Clasificación
La evaluación de modelos de clasificación se realiza mediante diversas métricas que permiten medir su rendimiento:

- **Precisión (Accuracy):** Proporción de predicciones correctas sobre el total de predicciones realizadas.

- **Precisión (Precision):** Indica la proporción de verdaderos positivos entre el total de positivos predichos, reflejando la exactitud de las predicciones positivas.

- **Exhaustividad (Recall o Sensibilidad):** Mide la proporción de verdaderos positivos detectados sobre el total de positivos reales, indicando la capacidad del modelo para identificar correctamente las instancias positivas.

- **Puntuación F1 (F1 Score):** Media armónica entre la precisión y la exhaustividad, proporcionando un balance entre ambas métricas.

La elección de la métrica adecuada depende del contexto y de la naturaleza del problema, especialmente en casos de desbalance de clases. [5]

#### Hiperparámetros
En aprendizaje automático, los hiperparámetros son configuraciones establecidas antes del proceso de entrenamiento que determinan el comportamiento del algoritmo y la estructura del modelo. A diferencia de los parámetros, que se aprenden durante el entrenamiento, los hiperparámetros controlan aspectos como la complejidad del modelo, la tasa de aprendizaje y el número de iteraciones. La selección adecuada de hiperparámetros es crucial para el rendimiento del modelo y, a menudo, se realiza mediante técnicas de búsqueda y validación cruzada. [6]

#### Optimización Bayesiana
La Optimización Bayesiana es una estrategia secuencial utilizada para encontrar el óptimo global de funciones de caja negra que son costosas de evaluar. Construye un modelo probabilístico del objetivo y utiliza este modelo para seleccionar los puntos más prometedores a evaluar, equilibrando la exploración y explotación del espacio de búsqueda. [7]

```python
def clasificar_region(ciudad):
    if any(x in ciudad for x in ["Pune", "Thane", "Nashik", "Vadodara", "Kalyan", "Rajkot", "Ahmedabad", "Mumbai", "Surat", "Vasai-Virar"]):
        return "Oeste"
    elif any(x in ciudad for x in ["Kolkata", "Patna"]):
        return "Este"
    elif any(x in ciudad for x in ["Srinagar", "Varanasi", "Jaipur", "Lucknow", "Ludhiana", "Meerut", "Agra", "Ghaziabad", "Kanpur", "Faridabad", "Delhi", "Khaziabad"]):
        return "Norte"
    elif any(x in ciudad for x in ["Nagpur", "Indore", "Bhopal"]):
        return "Centro"
    elif any(x in ciudad for x in ["Visakhapatnam", "Bangalore", "Chennai", "Hyderabad"]):
        return "Sur"
    else:
        return "Sin región"

data["Region"] = data["City"].apply(clasificar_region)
data.tail(3)
```

```python
data = data.drop(columns = ["id", "City"])
data["Sleep Duration"] = data["Sleep Duration"].str.extract(r'(\d+)').astype(float)
data["Family History of Mental Illness"] = data["Family History of Mental Illness"].map({'Yes': 1, 'No': 0})
data["Have you ever had suicidal thoughts ?"] = data["Have you ever had suicidal thoughts ?"].map({'Yes': 1, 'No': 0})
data["Gender"] = data["Gender"].map({'Male': 1, 'Female': 0})
data = data.dropna()

numerical_features = ["Age", "Academic Pressure", "Work Pressure", "Job Satisfaction", "CGPA", "Study Satisfaction", "Work/Study Hours", "Financial Stress", "Sleep Duration", "Family History of Mental Illness", "Have you ever had suicidal thoughts ?", "Gender"]
categorical_features = ["Profession", "Dietary Habits", "Degree", "Region"]

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

x = data.drop(columns = ["Depression"])
y = data["Depression"]
```

### Modelo Regresión Logística con Regularización Ridge.

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0)

gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
```

```python
def Regresion_Logis(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features):

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_features),
            ('cat', categorical_transformer, categorical_features)
        ]
    )

    model = LogisticRegression(penalty='l2', C=1.0, max_iter=1000)

    pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.4, random_state=2)

    search_space = {
        'model__C': Real(1e-2, 1e2, prior='log-uniform'),
    }

    opt = BayesSearchCV(
        estimator=pipeline,
        search_spaces=search_space,
        n_iter=20,
        scoring='roc_auc',
        cv=3,
        random_state=2,
        verbose=0,
        n_jobs=-1
    )

    opt.fit(X_train, y_train)

    best_model = opt.best_estimator_
    y_score = best_model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_score)
    auc_scores_cv = opt.cv_results_['mean_test_score']
    auc_var = np.var(auc_scores_cv)

    RocCurveDisplay.from_predictions(
        y_test,
        y_score,
        name="ROC Curve (Regresión Logídtica)",
        color="darkorange",
        plot_chance_level=True
    )
    plt.title("ROC curve (con optimización BayesSearchCV)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()

    print(f"Mejor C encontrado: {opt.best_params_['model__C']}")
    print(f"AUC final en test: {auc_score:.4f}")
    print(f"Varianza del AUC (en CV): {auc_var:.6f}")
```

```python
Regresion_Logis(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features)
```

### Máquina de Soporte con Kernel rbf.

```python
def Maquina_de_soporte_bayes(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features):

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_features),
            ('cat', categorical_transformer, categorical_features)
        ]
    )

    svc = SVC(kernel="rbf", probability=True)

    pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", svc)
    ])

    X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.4, random_state=2)

    search_space = {
        'model__C': Real(1e-2, 1e2, prior='log-uniform'),
        'model__gamma': Real(1e-3, 1, prior='log-uniform')
    }

    opt = BayesSearchCV(
        estimator=pipeline,
        search_spaces=search_space,
        n_iter=20,
        scoring='roc_auc',
        cv=3,
        random_state=2,
        verbose=0,
        n_jobs=-1
    )

    opt.fit(X_train, y_train)

    best_model = opt.best_estimator_
    y_score = best_model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_score)

    auc_scores_cv = opt.cv_results_['mean_test_score']
    auc_var = np.var(auc_scores_cv)


    RocCurveDisplay.from_predictions(
        y_test,
        y_score,
        name="ROC Curve (Maquina de Soporte)",
        color="darkorange",
        plot_chance_level=True
    )
    plt.title("ROC curve (con optimización BayesSearchCV)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()


    print(f"Mejor C encontrado: {opt.best_params_['model__C']}")
    print(f"Mejor gamma encontrado: {opt.best_params_['model__gamma']}")
    print(f"AUC final en test: {auc_score:.4f}")
    print(f"Varianza del AUC (en CV): {auc_var:.6f}")
```

```python
Maquina_de_soporte_bayes(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features)
```

### Rede Neuronal con dos capas ocultas.

```python
def Red_Neuronal(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features):

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_features),
            ('cat', categorical_transformer, categorical_features)
        ]
    )

    model = MLPClassifier(max_iter=1000, random_state=2)

    pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.4, random_state=2)

    search_space = {
        'model__hidden_layer_1': Integer(10, 100),
        'model__hidden_layer_2': Integer(10, 100),
        'model__alpha': Real(1e-5, 1e-1, prior='log-uniform')
    }

    class CustomMLPClassifier(MLPClassifier):
        def set_params(self, **params):
            if 'hidden_layer_1' in params or 'hidden_layer_2' in params:
                h1 = params.pop('hidden_layer_1', 50)
                h2 = params.pop('hidden_layer_2', 50)
                params['hidden_layer_sizes'] = (h1, h2)
            return super().set_params(**params)

    pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", CustomMLPClassifier(max_iter=1000, random_state=2))
    ])

    opt = BayesSearchCV(
        estimator=pipeline,
        search_spaces=search_space,
        n_iter=10,
        scoring='roc_auc',
        cv=3,
        random_state=2,
        n_jobs=-1
    )

    opt.fit(X_train, y_train)

    best_model = opt.best_estimator_
    y_score = best_model.predict_proba(X_test)[:, 1]
    auc_score = roc_auc_score(y_test, y_score)
    auc_scores_cv = opt.cv_results_['mean_test_score']
    auc_var = np.var(auc_scores_cv)

    RocCurveDisplay.from_predictions(
        y_test,
        y_score,
        name="ROC Curve (Red Neuronal)",
        color="darkorange",
        plot_chance_level=True
    )
    plt.title("ROC curve (con optimización BayesSearchCV)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.show()

    print(f"Mejor arquitectura de capas ocultas: {best_model.named_steps['model'].hidden_layer_sizes}")
    print(f"Mejor alpha (regularización): {opt.best_params_['model__alpha']}")
    print(f"AUC final en test: {auc_score:.4f}")
    print(f"Varianza del AUC (en CV): {auc_var:.6f}")
```

```python
Red_Neuronal(x, y, numerical_transformer, categorical_transformer, numerical_features, categorical_features)
```

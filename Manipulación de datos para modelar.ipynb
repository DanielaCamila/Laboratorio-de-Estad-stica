```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
```

```python
df
```

```python
df.describe()
```

Quitar los Nan

```python
df = df.dropna()
```

Este es un paso que se hace una vez que ya corriste por primera vez los datos y quitas las columnas que no son significativas

```python
df = df.drop(columns = ["AGE", "INDUS", "CRIM"])
```

Estamos escalando los datos

```python
for i in range(len(df.columns)):
    mini = df.iloc[:, i].min()
    dfmin = df.iloc[:, i] 
    maxi = dfmin.max()
    dfmax = dfmin/maxi
    df.iloc[:, i] = dfmax
```

```python
df.describe()
```

```python
import statsmodels.api as sm
```

```python
x = df.iloc[:, :-1].values
y = np.reshape(df["MEDV"].values, [-1,1])
n = len(y)
```

```python
ones = np.ones([n, 1])
X = np.hstack((ones, x))
ols = sm.OLS(df["MEDV"], X)
results = ols.fit()
results.summary()
```

Bootstrap con arboles

```python
data = pd.read_csv("Advertising.csv")

data = data.drop(columns = ["Unnamed: 0"])

data["TV_radio"] = data["TV"]*data["radio"]
data["TV_news"] = data["TV"]*data["newspaper"]
data["radio_news"] = data["radio"]*data["newspaper"]

data.head(3)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

factor = []
modelo = []
Y_pred = []

for i in range(1000):
    muestra = data.sample(n=len(data), replace=True)

    y = muestra["sales"].values 
    les = data.columns.drop("sales")

    factores_i = np.random.choice(les, size=4, replace=False)
    x = muestra[factores_i]

    factor.append(factores_i)

    model = LinearRegression()
    model_i = model.fit(x, y)
    modelo.append(model)

    y_pred_i = model_i.predict(x)
    Y_pred.append(y_pred_i)
```

```python
test = np.array(Y_pred)

#El promedio de las 1000 predicción por muestra
Promedio = test.mean(axis = 0)

r2 = r2_score(y, Promedio)
r2
```
